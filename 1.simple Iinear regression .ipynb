# defining x and y

```python
x = df[['YearsExperience']] # dataframe
y = df['Salary']  # This can stay as Series
```

# model traning

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x,y)
```

```python
lr.coef_ # slope = m
```

```python
lr.intercept_ # c
```

```python
print(f"slope is:{lr.coef_} ")
print(f"intercept is:{lr.intercept_} ")
```

```python
m= 9449.96232146
c= 24848.203966523208
x= 1.2

y= m*x+c
y
```

# predicting for actual value

```python
df
```

```python
lr.predict(df[['YearsExperience']])
```

```python
ypred =lr.predict(df[['YearsExperience']])
ypred
```

```python
y
```

# model evaluation

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mse=mean_squared_error(y,ypred)
rmse=np.sqrt(mse)
r2=r2_score(y,ypred)
```

```python
print(f'mse : {mse}')
print(f'rmes: {rmse}')
print(f'R2  : {r2}')
```
