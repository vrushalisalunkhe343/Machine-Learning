#  train_test_split,KFold,LeaveOneOut,cross_val_score

```python

import pandas as pd 
from sklearn.model_selection import train_test_split,KFold,LeaveOneOut,cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report,accuracy_score
```

```python
df=pd.read_csv(r'C:\Users\Sarvadnya\FUSIONCLASS\machine_learning\fetal_health.csv')
df
```

```python
df['fetal_health'].unique()
```

```python
# model building 
```

```python
x=df.iloc[:,:-1]
y=df['fetal_health']
```

```python
x
```

```python
y
```

```python
xtrain,ytrain,xtest,ytest=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
dt=DecisionTreeClassifier(criterion='entropy')
dt.fit(xtrain,ytrain)
yp =dt.predict(xtest)
yp
print(classification_report(y,yp))
```

```python
# k fold 
```

```python
kf=KFold(n_splits=5)
kf_accuracy=cross_val_score(DecisionTreeClassifier(),x,y,cv=kf).mean()
kf_accuracy
```
